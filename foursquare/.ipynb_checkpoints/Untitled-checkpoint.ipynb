```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
indata = pd.read_csv("checkins.dat", sep='|')
indata.head()
```

```python
data = indata.drop(indata.columns[[1,2,5]], 1)
data.replace('                   ', np.nan, inplace=True)
data.columns = data.columns.map(lambda x: x.strip())
data.dropna(inplace=True)
data = data.apply(pd.to_numeric)
data.head()
data.iloc[0,1]
```

```python
for f in [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5, 0.75, 1.0]:
    sample = data.sample(frac=f, random_state=1)
    plt.plot(sample['longitude'], sample['latitude'], '.')
    plt.savefig("img_"+str(f)+".jpg")
    plt.figure()
```
